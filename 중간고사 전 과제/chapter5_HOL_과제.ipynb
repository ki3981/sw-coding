### Task 3-2. `get_stats(numbers)` — 다중 반환

숫자 리스트를 받아 **(최솟값, 최댓값, 평균)**을 튜플로 반환하는 함수를 완성하세요.

```
[예상 출력]
데이터: [55, 82, 90, 73, 61, 47, 95, 68]
최솟값: 47
최댓값: 95
평균:   71.38
```

> 💡 힌트: `min()`, `max()`, `sum()`, `len()` 내장 함수를 활용하세요.


````python
def get_stats(numbers):
    # TODO: 최솟값, 최댓값, 평균을 계산하세요
    max_stats=max(numbers)
    min_stats=min(numbers)
    avg_stats=sum(numbers)/len(numbers)

    # TODO: return 으로 세 값을 동시에 반환하세요
    return min_stats,max_stats,avg_stats    #반환시 쉼표로 구분 필수

data = [55, 82, 90, 73, 61, 47, 95, 68]

# TODO: 반환된 세 값을 low, high, avg 변수에 각각 받으세요
low, high, avg = get_stats(data)
# TODO: 예상 출력 형식대로 출력하세요 (평균은 소수점 둘째 자리)
print(f'데이터: {data}')
print(f'최솟값: {low}')
print(f'최댓값: {high}')
print(f'평균:   {avg:.2f}')
````

## Lab 4. 전역 변수와 지역 변수

### 학습 목표
- 전역변수와 지역변수의 유효 범위(scope)를 구분할 수 있다.
- `global` 키워드를 올바르게 사용할 수 있다.

### 배경

| 구분 | 선언 위치 | 유효 범위 |
|------|-----------|-----------|
| **전역변수** (global variable) | 함수 바깥 | 프로그램 전체에서 읽기 가능 |
| **지역변수** (local variable) | 함수 내부 | 해당 함수 내부에서만 유효 |

```
score = 100          ← 전역변수
│
├── def func():
│       score = 999  ← 지역변수 (전역변수와 이름이 같아도 별개!)
│       print(score) # 999
│
print(score)         # 100 (전역변수 변화 없음)
```

전역변수를 함수 내부에서 **수정**하려면 반드시 `global`을 선언해야 합니다.

### Task 4-3. 심화 — 은행 잔액 관리

전역변수 `balance`를 이용해 입금/출금 함수를 작성하세요.

- `deposit(amount)` : `amount`만큼 잔액 증가 후 현재 잔액 출력
- `withdraw(amount)` : 잔액이 충분하면 차감, 부족하면 경고 메시지 출력

```
[예상 출력]
현재 잔액: 0 원
입금 50000 원 → 잔액: 50,000 원
입금 30000 원 → 잔액: 80,000 원
출금 20000 원 → 잔액: 60,000 원
출금 70000 원 → 잔액 부족! (현재 잔액: 60,000 원)
```


````python
balance = 0

def deposit(amount):
    # TODO: global balance 선언 후 amount만큼 증가시키고 잔액을 출력하세요
    global balance
    balance += amount
    print(f'입금 {amount:,} 원 -> 잔액: {balance:,} 원')

def withdraw(amount):
    # TODO: global balance 선언 후
    global balance

    #잔액이 충분하면 amount를 차감하고 출력하세요
    if balance >= amount:
        balance -= amount
        print(f'출금 {amount:,} 원 -> 잔액: {balance:,} 원')
    #잔액이 부족하면 경고 메시지를 출력하세요
    else: 
        print(f'출금 {amount:,} 원 -> 잔액부족! (현재 잔액: {balance:,} 원)')
    

deposit(50000)
deposit(30000)
withdraw(20000)
withdraw(70000)
````

## Lab 6. 재귀함수 (Recursion)

### 학습 목표
- 재귀함수의 구조(종료 조건 + 재귀 호출)를 이해할 수 있다.
- 재귀로 수학적 문제를 간결하게 표현할 수 있다.

### 배경

재귀함수는 **자기 자신을 호출**하는 함수입니다.
반드시 두 가지 요소가 있어야 합니다.

```
① 종료 조건 (Base Case)      → 재귀를 멈추는 조건
② 재귀 호출 (Recursive Call) → 자기 자신을 더 작은 문제로 호출
```

```python
def countdown(n):
    if n == 0:           # ① 종료 조건
        print("발사!")
        return
    print(n)
    countdown(n - 1)     # ② 재귀 호출 (n이 점점 작아짐)

countdown(3)
# 출력: 3 → 2 → 1 → 발사!
```

> ⚠️ 종료 조건이 없으면 **무한 재귀(RecursionError)**가 발생합니다.


### Task 6-1. `power(base, exp)` — 거듭제곱 (재귀)

`base`의 `exp` 제곱을 재귀로 계산하세요.

```
수학적 정의:
  power(base, 0)   = 1                       ← 종료 조건
  power(base, exp) = base × power(base, exp-1)  ← 재귀
```

```
[예상 출력]
2^0 = 1
2^5 = 32
3^4 = 81
```


````python
def power(base, exp):
    # TODO: 종료 조건 — exp가 0이면 1을 반환하세요
    if exp == 0:
        return 1
    # TODO: 재귀 호출 — base * power(base, exp-1)을 반환하세요
    else:
        return base * power(base, exp-1)



print(f"2^0 = {power(2, 0)}")
print(f"2^5 = {power(2, 5)}")
print(f"3^4 = {power(3, 4)}")
````

### Task 6-2. `digit_sum(n)` — 자릿수 합 (재귀)

정수 `n`의 각 자릿수를 모두 더한 값을 재귀로 구하세요.

```
예: digit_sum(1234) = 4 + digit_sum(123)
                    = 4 + 3 + digit_sum(12)
                    = 4 + 3 + 2 + digit_sum(1)
                    = 4 + 3 + 2 + 1 = 10
```

```
[예상 출력]
digit_sum(1234) = 10
digit_sum(9876) = 30
digit_sum(100)  = 1
```

> 💡 힌트: `n % 10` → 1의 자리 숫자 / `n // 10` → 나머지 자리


````python
def digit_sum(n):
    # TODO: 종료 조건 — n이 한 자리 수(n < 10)이면 n을 반환하세요
    if n < 10:
        return n
    # TODO: 재귀 호출 — n % 10 + digit_sum(n // 10)을 반환하세요
    else: 
        return n % 10 + digit_sum(n // 10)
    

print(f"digit_sum(1234) = {digit_sum(1234)}")
print(f"digit_sum(9876) = {digit_sum(9876)}")
print(f"digit_sum(100)  = {digit_sum(100)}")
````

### Task 6-3. 심화 — `hanoi()` 하노이 탑 (재귀)

`n`개의 원판을 `from_peg`에서 `to_peg`로, `via_peg`를 경유하여 옮기는 모든 단계를 출력하세요.

```
[예상 출력 — hanoi(3, 'A', 'C', 'B') 호출 시]
A → C
A → B
C → B
A → C
B → A
B → C
A → C
```

> 💡 알고리즘 (3단계):
> 1. 위 `n-1`개를 `from_peg → via_peg` (`to_peg` 경유)로 옮긴다
> 2. 가장 큰 원판을 `from_peg → to_peg`로 옮긴다 (출력)
> 3. `n-1`개를 `via_peg → to_peg` (`from_peg` 경유)로 옮긴다


````python
def hanoi(n, from_peg, to_peg, via_peg):
    # TODO: 종료 조건 — n이 1이면 "from_peg → to_peg" 출력 후 반환하세요
    if n == 1:
        print(f"{from_peg} -> {to_peg}")
        return
    # TODO: 재귀 1단계 — 위 n-1개를 from_peg에서 via_peg로 (to_peg 경유)
    hanoi(n-1, from_peg, via_peg, to_peg)

    # TODO: 가장 큰 원판 이동 출력
    print(f"{from_peg} -> {to_peg}")
    # TODO: 재귀 2단계 — n-1개를 via_peg에서 to_peg로 (from_peg 경유)
    hanoi(n-1, via_peg, to_peg, from_peg)
    

hanoi(3, 'A', 'C', 'B')
````

### Task 8-3. 심화 — `safe_sqrt()` 입력 검증

`safe_sqrt(value)` 함수를 작성하세요.

- 숫자(int 또는 float)이고 0 이상이면 제곱근을 반환합니다.
- 음수이면 `"음수는 제곱근 계산 불가"` 문자열을 반환합니다.
- 숫자가 아니면 `"숫자를 입력하세요"` 문자열을 반환합니다.

```
[예상 출력]
safe_sqrt(16)      → 4.0
safe_sqrt(2)       → 1.4142135623730951
safe_sqrt(-9)      → 음수는 제곱근 계산 불가
safe_sqrt("hello") → 숫자를 입력하세요
```

> 💡 `type(value) in (int, float)` 또는 `isinstance(value, (int, float))` 사용


````python
import math

def safe_sqrt(value):
    # TODO: type()으로 int 또는 float인지 확인하세요
    #       숫자가 아니면 "숫자를 입력하세요" 반환
    if type(value) != int and type(value) != float:
        return "숫자를 입력하세요"
    # TODO: 음수이면 "음수는 제곱근 계산 불가" 반환
    elif value < 0:
        return "음수는 제곱근 계산 불가"
    # TODO: 조건을 모두 통과하면 math.sqrt(value) 반환
    else:
        return math.sqrt(value)
    
    

print(f"safe_sqrt(16)      → {safe_sqrt(16)}")
print(f"safe_sqrt(2)       → {safe_sqrt(2)}")
print(f"safe_sqrt(-9)      → {safe_sqrt(-9)}")
print(f"safe_sqrt('hello') → {safe_sqrt('hello')}")
````

## 🏆 종합 프로젝트 — 학생 성적 관리 시스템

### 설명
지금까지 배운 **모든 개념**을 활용하여 성적 관리 시스템을 완성하세요.

### 구현할 함수

| 함수 | 설명 | 활용 개념 |
|------|------|-----------|
| `add_student(name, *scores)` | 학생을 전역 리스트에 추가 | 가변 인자, 전역변수 |
| `get_grade(avg, a=90, b=80, c=70, d=60)` | 평균으로 학점 반환 | 디폴트 인자, return |
| `calc_stats(scores)` | (합계, 평균, 최고, 최저) 튜플 반환 | 다중 반환, 내장 함수 |
| `print_report(show_detail=True)` | 전체 성적 리포트 출력 | 디폴트 인자, f-문자열 |

### 예상 출력

```
============================================================
                    📋 학생 성적 리포트
============================================================
 번호  이름            합계    평균   최고   최저  학점
------------------------------------------------------------
    1  김파이썬          255    85.0     92     78    B
    2  이자바            198    66.0     78     55    D
    3  박씨쁠쁠          280    93.3    100     88    A
------------------------------------------------------------
  전체 평균: 81.4점
============================================================
```


````python
# 전역 리스트 — 학생 데이터 저장소
students = []

def add_student(name, *scores):
    # TODO: {"name": name, "scores": list(scores)} 형태로
    #       students 리스트에 추가하세요
    for score in scores:
        if score < 0 or score > 100:
            print(f"경고: {name}의 점수 {score}는 0~100 범위를 벗어났습니다.")
            return
    students.append((name,list(scores)))
    

def get_grade(avg, a=90, b=80, c=70, d=60):
    # TODO: avg가 a 이상이면 "A", b 이상이면 "B", ...
    #       디폴트 인자를 이용해 학점 기준을 받고 해당 학점 문자를 반환하세요
    if avg >= a:
        return "A"
    elif avg >= b:
        return "B"
    elif avg >= c:
        return "C"
    elif avg >= d:
        return "D"
    else:
        return "F"
    

def calc_stats(scores):
    # TODO: sum(), len(), max(), min()을 활용하여
    #       합계, 평균, 최고점, 최저점을 계산하고 튜플로 반환하세요
    total = sum(scores)
    avg = total / len(scores)
    highest = max(scores)
    lowest = min(scores)
    return total, avg, highest, lowest
    

def print_report(show_detail=True):
    # TODO: "===..." 헤더와 제목을 출력하세요
    print("=" * 60)
    print(f'{"📋 학생 성적 리포트":>30}')
    print("=" * 60)
    if show_detail:
        print(f'{"번호":<4} {"이름":<8} {"합계":>5} {"평균":>6} {"최고":>5} {"최저":>5} {"학점":>4}')
        print("-" * 60)
    # TODO: show_detail이 True이면 헤더 행과 각 학생 행을 출력하세요
    #       (번호, 이름, 합계, 평균, 최고, 최저, 학점)
    avg_list = []
    for i, (name, scores) in enumerate(students, start=1):
        total, avg, highest, lowest = calc_stats(scores)
        grade = get_grade(avg)
        avg_list.append(avg)
        if show_detail :
            print(f'{i:<4} {name:<8} {total:>5} {avg:>7.1f} {highest:>7} {lowest:>7} {grade:>4}')


    # TODO: 전체 평균을 계산하여 하단에 출력하세요
    print("-" * 60)
    print(f"전체 평균: {sum(avg_list) / len(avg_list):.1f}점")
    print("=" * 60)


def recursive_sort_students(student_list):
    if len(student_list) <= 1:
        return student_list

    pivot = student_list[0]
    pivot_avg = calc_stats(pivot[1])[1]

    higher = []
    lower = []

    for student in student_list[1:]:
        student_avg = calc_stats(student[1])[1]
        if student_avg > pivot_avg:
            higher.append(student)
        else:
            lower.append(student)

    return recursive_sort_students(higher) + [pivot] + recursive_sort_students(lower)


# ── 실행 ──────────────────────────────────────────────────
add_student("김파이썬",  85, 92, 78)
add_student("이자바",    55, 65, 78)
add_student("박씨쁠쁠",  88, 100, 92)
add_student("최에러",    120, 80, 90)

students = recursive_sort_students(students)

print_report()
print_report(show_detail=False)
````
