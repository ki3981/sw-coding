# Practice 3.11

```python
a, b, c = map(int, input("세 복권번호를 입력하세요 : ").split())

my_nums = [a, b, c]
win_nums = [2, 3, 9]

count = 0

for n in my_nums:
    if n in win_nums:
        count += 1

if count == 3:
    print("1억 원")
elif count == 2:
    print("1천만 원")
elif count == 1:
    print("1만 원")
else:
    print("다음 기회에")
```

# Practice 4.4

```python
print("맛나 식당에 오신 것을 환영합니다. 메뉴는 다음과 같습니다.")
print("1) 햄버거")
print("2) 치킨")
print("3) 피자")

menu = int(input("1부터 3까지의 메뉴를 선택하세요 : "))

while menu < 1 or menu > 3:
    print("메뉴를 다시 선택하세요")
    menu = int(input("1부터 3까지의 메뉴를 선택하세요 : "))

if menu == 1:
    print("햄버거를 선택하셨습니다.")
elif menu == 2:
    print("치킨을 선택하셨습니다.")
elif menu == 3:
    print("피자를 선택하셨습니다.")
```

# Practice 4.8

```python
for i in range(2, 13):   # 2부터 12까지
    is_prime = True

    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break

    if is_prime:
        print(i, ": 소수")
    else:
        print(i, ": 합성수")
```
