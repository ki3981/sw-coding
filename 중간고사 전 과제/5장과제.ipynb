## LAB 5-6

```python
def print_root(a, b, c):
    r1 = (-b + (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)
    r2 = (-b - (b ** 2 - 4 * a * c) ** 0.5) / (2 * a)

    print("해는", r1, "또는", r2)

# (1) x^2 + 4x - 21 = 0
print_root(1, 4, -21)

# (2) x^2 - 6x + 8 = 0
print_root(1, -6, 8)
```

## LAB 5-7

```python
def circle_area_circum(radius):
    area = 3.14 * radius * radius
    circum = 2 * 3.14 * radius
    return area, circum

radius = 10
area, circum = circle_area_circum(radius)

print("반지름 10인 원의 면적은", area, ", 원의 둘레는", circum)
```

## LAB 5-8

```python
def multiples(n, m):
    result = []
    for i in range(1, m + 1):
        result.append(n * i)
    return tuple(result)

r1, r2, r3, r4 = multiples(3, 4)
print(r1, r2, r3, r4)


r1, r2, r3, r4, r5= multiples(2, 5)
print(r1, r2, r3, r4, r5)

    
```

## 연습문제 5.8

```python
def mean3(a, b, c):
    return (a + b + c) / 3

def max3(a, b, c):
    return max(a, b, c)

def min3(a, b, c):
    return min(a, b, c)

def mean6(a, b, c, d, e, f):
    return (a + b + c + d + e + f) / 6

def max6(a, b, c, d, e, f):
    return max(max3(a, b, c), max3(d, e, f))

def min6(a, b, c, d, e, f):
    return min(min3(a, b, c), min3(d, e, f))

a, b, c, d, e, f = map(int, input("여섯 개의 수를 입력하시오 : ").split())

print("평균값은", mean6(a, b, c, d, e, f))
print("최댓값은", max6(a, b, c, d, e, f))
print("최솟값은", min6(a, b, c, d, e, f))
```
