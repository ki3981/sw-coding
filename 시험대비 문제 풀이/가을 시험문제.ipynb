## 시험 1번 

```python
def calculate_display(diagnal_inch,aspect_ratio='16:9'):
    W,H = map(int,aspect_ratio.split(':'))
    if diagnal_inch < 10 or diagnal_inch > 100 :
        print('Invalid screen size')
        return None
    elif not ((W == 16 and H == 9) or (W == 16 and H == 10) or (W == 21 and H == 9) or (W == 4 and H == 4)):
        print('Invalid aspect size')
        return None  
    else:
        diagnal_cm = round(diagnal_inch * (2.54),2)
    
        width_cm = round((diagnal_cm * W)/(W**2+H**2)**0.5,2)
        height_cm = round(diagnal_cm * H/(W**2+H**2)**0.5,2)
        area_cm2 = round(width_cm * height_cm,2)
        viewing_distance_cm = round(area_cm2 * 1.5,2)
        return (diagnal_cm,width_cm,height_cm,area_cm2,viewing_distance_cm)

calculate_display(10,'16:9')
```

```python
def collatz_steps(n):
    
    if n < 1:
        print('input value is not valid')
        return None
    else :
        step = 0
        while n !=1 :
            
            if n % 2 == 0:
                n = n/2
                
            else :
                n= 3*n+1
            step +=1
        return step
    
collatz_steps(6)


def find_max_step(start, end):
    max_step = collatz_steps(start)
    max_step_num = start

    for i in range(start, end + 1):
        steps = collatz_steps(i)

        if steps > max_step:
            max_step = steps
            max_step_num = i
        elif steps == max_step and i < max_step_num:
            max_step_num = i

    return max_step_num, max_step


find_max_step(2,7)
```
