## 맴버연산자 in not in

```python
nation=['korea','china','russia','malaysia']
nation.append('Nepal')

print(nation)

if 'japan' not in nation == True:
    print('japan은 국가 목록에 없습니다.')
else:
    print('russia는 국가 목록에 없습니다.')
```

## string을 받아서 vowel을 삭제하기

```python
string = list(input())

for i in string[:]:
    if i in ['a','e','i','o','u']:
        string.remove(i)
print(string)
        
```
