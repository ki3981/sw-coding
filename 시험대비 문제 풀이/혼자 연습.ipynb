## 리스트 5.21 ## 다시풀기

```python
src='a2b3c6a3f1'
src1=list(src)
h=[]
for i in range(len(src)):
    if i % 2 == 0:
        h.append(src1[i])
        h.append(src1[i]*(int(src1[i+1])-1))
print(''.join(h))
print(h)       
```

## 3.20 이중 for 문  ## 다시풀기

```python
a = int(input('숫자를 입력하세요: '))

for i in range(1,a+1):
    for j in range(a):
        if j < a - i:
            print(' ', end='')
        else:
            print('*', end='')
    print()
```

## 3.23 연습문제

```python
a = int(input('숫자를 입력하세요: '))
total = 0
for i in range(1,1+a):
    total += i ** 2
print(total)
```
