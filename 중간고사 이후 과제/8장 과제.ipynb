## LAB 8-1


* `datetime.now()`로 현재 날짜/시간 객체를 생성해 `now`에 저장
* `now.year`, `now.month`, `now.day`로 오늘 날짜를 출력
* `now.hour < 12` 조건으로 오전/오후를 판별해 12시간 형식으로 시간 출력

```python
from datetime import datetime
now = datetime.now()
print(f'오늘의 날짜 : {now.year}년 {now.month}월 {now.day}일')

if now.hour < 12:
    a = '오전'
    hour = now.hour
else:
    a = '오후'
    hour = now.hour - 12
print(f'현재시간 : {a} {hour}시 {now.minute}분 {now.second}초')
```

## LAB 8-2

* `datetime(년, 월, 일)`로 목표 날짜 객체를 생성

* `gap.days`로 남은 일수, `gap.seconds // 3600`으로 남은 시간 출력

```python
from datetime import datetime

today = datetime.now()
print(f'오늘은 {today.year}년 {today.month}월 {today.day}일입니다.')

xmas = datetime(2026, 12, 25)
gap = xmas - today
print(f'2027년 크리스마스 까지는 {gap.days}일 {gap.seconds // 3600}시간 남았습니다.')

```

## LAB 8-6

**문제 1.** 코드 8-8에 `for _ in range(5)` 반복문을 추가해 로또 번호 5세트를 생성하여 출력

**문제 2.** `random.sample()`로 당첨 번호 6개와 보너스 번호 1개를 각각 추출하여 출력

```python
import random as rd

print('이번 주의 추천 로또 번호')
for i in range(5):                    # 5번 반복 (수정된 부분)
    lotto_list = list(range(1, 46))   # 1부터 45까지 생성
    rd.shuffle(lotto_list)            # 임의의 순서로 섞기
    lotto_list = lotto_list[:6]       # 앞 부분 6개만 선택
    lotto_list.sort()                 # 선택된 번호를 정렬
    print(lotto_list)
```

```python
import random

lotto = random.sample(range(1, 46), 6)
lotto.sort()
bonus = random.sample(range(1, 46), 1)

print('이번 주의 당첨 로또 번호 + 보너스')
print(f'{lotto} + {bonus}')
```

## 연습문제 8.1 - 춘향이와 몽룡이의 연애 기념일

**(1)** `date.today()`로 오늘 날짜를 가져와 연애 시작일(2019.02.14)과의 차이를 `timedelta`로 계산하여 경과 일수 출력

**(2)** `timedelta(days=N)`으로 100일, 200일, 500일, 1000일 기념일을 각각 계산하여 출력

```python

from datetime import date

start_day = date(2019, 2, 14)
today = date.today()

passed_days = today - start_day

print(f"춘향이와 몽룡이의 연애 시작일 : {start_day.year}년 {start_day.month}월 {start_day.day}일")
print(f"연애 시작일로부터 경과한 날짜 : {passed_days.days}일")
```

```python

from datetime import date, timedelta

start_day = date(2019, 2, 14)

print(f"춘향이와 몽룡이의 연애 시작일 : {start_day.year}년 {start_day.month}월 {start_day.day}일")
print()

anniversary_day = [100, 200, 500, 1000]

for day_count in anniversary_day:
    anniversary = start_day + timedelta(days=day_count)

    print(
        f"{day_count:,}일 기념일 : "
        f"{anniversary.year}년 {anniversary.month}월 {anniversary.day}일"
    )
```

## 연습문제 8.2 - 1부터 1,000,000까지의 합 계산 시간 측정

**(1)** `time.time()`으로 시작/종료 시각을 기록하여 `sum1to1000000()` 함수 1회 실행 시간을 초 단위로 출력

**(2)** `for`문으로 `sum1to1000000()` 함수를 100번 반복 호출하여 총 소요 시간을 출력

```python
import time

def sum1to1000000():
    total = 0

    for i in range(1, 1000001):
        total += i

    return total


start_time = time.time()

result = sum1to1000000()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"1에서 1,000,000까지의 합을 구하는 시간 : {elapsed_time:.4f}초")
```

## 연습문제 8.3 - 함수 100번 수행 시간 측정

**(1)** `"Hello Python!"`을 출력하는 함수를 100번 반복 수행하여 소요 시간을 출력

**(2)** `for`문으로 1~10000을 곱해 팩토리얼을 구하는 함수를 100번 반복 수행하여 소요 시간을 출력

**(3)** `range(1, 10001, 2)`로 홀수만 추출해 세제곱 합을 구하는 함수를 100번 반복 수행하여 소요 시간을 출력

**(4)** `math.radians()`와 `math.sin()`으로 1~360도의 sin값 합을 구하는 함수를 100번 반복 수행하여 소요 시간을 출력

### (1) 'Hello Python!'을 출력하는 함수

```python
import time

def hello_python():
    print("Hello Python!")


start_time = time.time()

for i in range(100):
    hello_python()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"Hello Python! 출력을 100번 반복하는 시간 : {elapsed_time:.4f}초")
```

### (2) 10000 팩토리얼을 구하는 함수

```python
import time

def factorial_10000():
    result = 1

    for i in range(1, 10001):
        result *= i

    return result


start_time = time.time()

for i in range(100):
    result = factorial_10000()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"10000!을 100번 반복해서 구하는 시간 : {elapsed_time:.4f}초")
```

### (3) 1에서 10,000까지의 모든 홀수를 세제곱하여 더한 값을 반환하는 함수

```python
import time

def sum_odd_cubes():
    total = 0

    for i in range(1, 10001, 2):
        total += i ** 3

    return total


start_time = time.time()

for i in range(100):
    result = sum_odd_cubes()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"1에서 10000까지 홀수의 세제곱 더하기를 100번 반복하는 시간 : {elapsed_time:.4f}초")
```

### (4) 1도에서 360도 사이의 sin값을 1도 단위로 구하여 그 합을 반환하는 함수

```python
import time
import math

def sum_sin_values():
    total = 0

    for degree in range(1, 361):
        radian = math.radians(degree)
        total += math.sin(radian)

    return total


start_time = time.time()

for i in range(100):
    result = sum_sin_values()

end_time = time.time()

elapsed_time = end_time - start_time

print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {elapsed_time:.4f}초")
```

## 연습문제 8.4 - 의사난수 함수 my_rand() 생성

**(1)** `X_{n+1} = (1103515245 * X_n + 12345) % 2^31` 공식으로 seed를 갱신하여 0~1,000,000 사이의 의사난수 10개를 출력

**(2)** `int(time.time() * 1000)`으로 초기 seed를 설정하여 매 실행마다 다른 난수가 생성되도록 하고 5개씩 2번 반복 출력

```python
seed = 123456

def my_rand():
    global seed

    num1 = 1103515245
    num2 = 12345
    num3 = 2 ** 31

    seed = (num1 * seed + num2) % num3

    return seed % 1000001


for i in range(10):
    print(my_rand())
```

```python
import time

seed = int(time.time() * 1000)

def my_rand():
    global seed

    num1 = 1103515245
    num2 = 12345
    num3 = 2 ** 31

    seed = (num1 * seed + num2) % num3

    return seed % 1000001

for i in range(2):
    for j in range(5):
        print(my_rand())
    print()
```
