# 🍳 Sequential Statements (순차 구조)
### — 요리 레시피로 배우는 프로그램의 흐름 —

> **핵심 개념**: 컴퓨터는 코드를 **위에서 아래로, 한 줄씩** 실행합니다.  
> 마치 요리 레시피를 **1번 → 2번 → 3번** 순서대로 따라 하는 것과 같습니다.

---

## 📌 오늘 배울 세 가지 제어 구조 미리 보기

| 구조 | 레시피 비유 | Python |
|------|------------|--------|
| **Sequential** | 1번, 2번, 3번… 순서대로 조리 | 위 → 아래 순서 실행 |
| **Conditional** | "간이 부족하면 소금을 더 넣는다" | `if / elif / else` |
| **Loop** | "3분 동안 계속 저어준다" | `for / while` |

> 오늘은 이 중 **Sequential (순차 구조)** 를 집중적으로 살펴봅니다.


---
## 1. 🍜 요리 레시피 = 순차 구조

라면 끓이는 순서를 생각해 봅시다.

```
① 냄비에 물 500ml를 붓는다
② 물이 끓을 때까지 가열한다
③ 스프를 넣는다
④ 면을 넣는다
⑤ 3분 더 끓인다
⑥ 그릇에 담아 완성!
```

> ⚠️ 만약 ④를 먼저 하고 ②를 한다면?  
> → 차가운 물에 면이 들어가 면이 불어버립니다!  
> **순서가 결과를 결정합니다.**


````python
# 🍜 라면 끓이기 — Python으로 표현한 순차 구조

print("=== 라면 끓이기 시작 ===")

# ① 재료 준비 (변수 선언)
water_ml   = 500
soup_pack  = 1
noodle_g   = 100

print(f"① 물 {water_ml}ml 준비 완료")
print(f"   스프 {soup_pack}봉지, 면 {noodle_g}g 준비 완료")

# ② 물 끓이기
is_boiling = True
print("② 물을 끓이는 중... 🔥")

# ③ 스프 투입
remaining_soup = soup_pack - 1
print(f"③ 스프 투입 완료! (남은 스프: {remaining_soup}봉지)")

# ④ 면 투입
remaining_noodle = noodle_g - noodle_g  # 전부 사용
print(f"④ 면 투입 완료! (남은 면: {remaining_noodle}g)")

# ⑤ 조리 시간 계산
cook_time_min = 3
print(f"⑤ {cook_time_min}분 더 끓이는 중... ⏳")

# ⑥ 완성
total_calories = 500
print(f"⑥ 완성! 🍜  (약 {total_calories}kcal)")
print("=== 라면 완성! ===")
````

---
## 2. ❌ 순서가 바뀌면 어떻게 될까?

변수를 **선언하기 전에** 사용하면 Python은 오류를 냅니다.  
요리에서 재료가 없는데 요리하려는 것과 같습니다!


````python
# ❌ 잘못된 순서 예시 — 실행하면 NameError 발생!

# total을 계산하려는데 아직 변수가 정의되지 않았음
try:
    total_cost = water_price + noodle_price   # 아직 선언 안 됨!
    print(total_cost)
except NameError as e:
    print(f"🚨 오류 발생: {e}")
    print("   → 레시피대로 재료를 먼저 준비해야 합니다!")

# ✅ 올바른 순서
water_price  = 50     # 먼저 선언
noodle_price = 800
total_cost   = water_price + noodle_price   # 그 다음 계산
print(f"✅ 올바른 순서: 재료비 합계 = {total_cost}원")
````

---
## 3. 🔄 변수는 요리 도중 바뀌는 재료량

요리하면서 재료가 줄어들 듯, 변수도 코드가 진행되면서 값이 바뀝니다.  
Python은 **가장 마지막에 대입된 값**을 기억합니다.


````python
# 🥘 찌개 끓이기 — 재료가 줄어드는 과정 추적

stock_ml = 1000
print(f"[시작]   육수: {stock_ml}ml")

# 두부 넣고 한 번 줄이기
stock_ml = stock_ml - 100   # 증발
print(f"[두부 투입 후] 육수: {stock_ml}ml  (100ml 증발)")

# 버섯 넣고 또 줄이기
stock_ml = stock_ml - 50
print(f"[버섯 투입 후] 육수: {stock_ml}ml  (50ml 증발)")

# 물 보충
stock_ml = stock_ml + 200
print(f"[물 보충 후]   육수: {stock_ml}ml  (200ml 추가)")

print(f"최종 육수량: {stock_ml}ml")
````

## 4. 🧁 실전 예제: 케이크 레시피 계산기

순차 구조를 활용해 케이크 재료비와 1인분 칼로리를 계산해 봅니다.


````python
# 🎂 케이크 레시피 계산기

print("=" * 40)
print("   🎂 초콜릿 케이크 레시피 계산기")
print("=" * 40)

# ── 재료 & 가격 ──────────────────────
flour_g      = 200;   flour_price    = 300   # 밀가루 200g
sugar_g      = 150;   sugar_price    = 200   # 설탕 150g
butter_g     = 100;   butter_price   = 800   # 버터 100g
chocolate_g  = 80;    chocolate_price= 1200  # 초콜릿 80g
egg_count    = 3;     egg_price      = 300   # 달걀 3개

print("📋 재료 목록")
print(f"  밀가루   {flour_g}g    ···  {flour_price}원")
print(f"  설탕     {sugar_g}g    ···  {sugar_price}원")
print(f"  버터     {butter_g}g   ···  {butter_price}원")
print(f"  초콜릿   {chocolate_g}g ···  {chocolate_price}원")
print(f"  달걀     {egg_count}개  ···  {egg_price}원")

# ── 총 재료비 계산 ────────────────────
total_cost = flour_price + sugar_price + butter_price + chocolate_price + egg_price
print(f"💰 총 재료비: {total_cost}원")

# ── 칼로리 계산 ───────────────────────
cal_flour     = flour_g     * 3.6   # 밀가루 3.6kcal/g
cal_sugar     = sugar_g     * 4.0   # 설탕   4.0kcal/g
cal_butter    = butter_g    * 7.2   # 버터   7.2kcal/g
cal_chocolate = chocolate_g * 5.5   # 초콜릿 5.5kcal/g
cal_egg       = egg_count   * 78    # 달걀   78kcal/개

total_cal   = cal_flour + cal_sugar + cal_butter + cal_chocolate + cal_egg
servings    = 8
cal_per_srv = total_cal / servings

print(f"🔥 총 칼로리: {total_cal:.1f} kcal")
print(f"🍰 1조각({servings}등분) 칼로리: {cal_per_srv:.1f} kcal")
print("✅ 케이크 완성! 🎂")
````

## 5. ✅ 핵심 정리

| 포인트 | 설명 |
|--------|------|
| **위 → 아래** | Python은 코드를 위에서 아래로 한 줄씩 실행 |
| **순서가 결과를 결정** | 선언 전 변수 사용 → `NameError` |
| **변수는 덮어쓰기 가능** | 같은 변수에 새 값을 대입하면 이전 값은 사라짐 |
| **다음 단계 예고** | 조건(`if`)과 반복(`for/while`)으로 흐름을 제어하는 법을 배울 예정 |

> 🍳 **"좋은 코드는 좋은 레시피처럼, 누가 읽어도 순서가 명확해야 합니다."**
