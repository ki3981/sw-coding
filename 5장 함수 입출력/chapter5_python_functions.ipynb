## 5.1 함수의 역할

### 왜 함수가 필요한가?

아래처럼 같은 코드를 반복해서 쓰는 상황을 생각해 봅시다.


````python
# ❌ 함수 없이 반복 코드 작성
print("=" * 20)
print("   환영합니다!")
print("=" * 20)

print("=" * 20)
print("   환영합니다!")
print("=" * 20)

print("=" * 20)
print("   환영합니다!")
print("=" * 20)
````

같은 코드가 3번 반복됩니다. 만약 출력 형태를 바꾸고 싶다면 3곳을 모두 수정해야 합니다.  
**함수**를 사용하면 이를 깔끔하게 해결할 수 있습니다.


````python
# ✅ 함수를 사용한 코드 — 한 번 정의, 여러 번 호출
def show_banner():
    print("=" * 20)
    print("   환영합니다!")
    print("=" * 20)

show_banner()  # 1번째 호출
show_banner()  # 2번째 호출
show_banner()  # 3번째 호출
````

````python
# 내장 함수 예시
numbers = [5, 3, 8, 1, 9, 2]

print("리스트:", numbers)
print("길이:", len(numbers))
print("최솟값:", min(numbers))
print("최댓값:", max(numbers))
print("정렬:", sorted(numbers))
````

## 5.2 함수와 매개변수

### 매개변수(parameter)와 인자(argument)

- **매개변수**: 함수 정의 시 괄호 안의 변수
- **인자**: 함수 호출 시 전달하는 값

```
def 함수이름(매개변수):   ← 매개변수
    ...

함수이름(인자)            ← 인자
```


````python
# 단일 매개변수 — 별표 줄 수 제어
def print_star(n):
    for _ in range(n):
        print("★" * 10)

print_star(1)
print()
print_star(3)
print()
print_star(5)
````

### 다중 매개변수

두 개 이상의 매개변수를 사용할 수 있습니다.  
호출 시 **인자의 수**가 매개변수의 수와 반드시 일치해야 합니다.


````python
# 다중 매개변수 — 도형 출력기
def draw_shape(symbol, width, height):
    for _ in range(height):
        print(symbol * width)

draw_shape("■", 8, 3)
print()
draw_shape("●", 5, 4)
````

````python
# TypeError 예시 — 인자 수 불일치
def add(a, b):
    return a + b

try:
    add(1)  # 인자가 1개뿐 → TypeError 발생
except TypeError as e:
    print(f"오류 발생: {e}")

print(add(3, 5))  # ✅ 올바른 호출
````

## 5.3 실습 — 2차 방정식의 근 구하기

### 함수 없이 작성한 버전 (문제점 확인)


````python
import math

# ❌ 함수 없이 — 계수가 바뀔 때마다 코드를 복사해야 함
a, b, c = 1, -3, 2
r1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
r2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)
print(f"x² - 3x + 2 = 0  →  r1={r1}, r2={r2}")

a, b, c = 2, -6, -8
r1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
r2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)
print(f"2x² - 6x - 8 = 0  →  r1={r1}, r2={r2}")
````

### 함수를 사용한 버전 (개선)

매개변수 a, b, c를 받아 근의 공식을 한 번만 작성하면 됩니다.


````python
import math

def print_root(a, b, c):
    """ax² + bx + c = 0의 근을 출력하는 함수"""
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        print("실수 범위에서 근이 없습니다.")
        return
    r1 = (-b + math.sqrt(discriminant)) / (2*a)
    r2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"방정식의 근: r1 = {r1:.2f}, r2 = {r2:.2f}")

print_root(1, -3, 2)    # x² - 3x + 2 = 0
print_root(2, -6, -8)   # 2x² - 6x - 8 = 0
print_root(1, 1, 1)     # 근 없음
````

## 5.4 return을 이용한 결과값 반환

### 단일 반환값


````python
def celsius_to_fahrenheit(c):
    """섭씨를 화씨로 변환"""
    return c * 9 / 5 + 32

temp_c = 100
temp_f = celsius_to_fahrenheit(temp_c)
print(f"{temp_c}°C = {temp_f}°F")

# return 결과를 바로 계산에 활용
print(f"체온 36.5°C = {celsius_to_fahrenheit(36.5):.1f}°F")
````

### 다중 반환값 (튜플 반환)

쉼표로 구분하면 여러 값을 한 번에 반환할 수 있습니다.  
반환된 값은 **튜플(tuple)** 형태로 전달됩니다.


````python
import math

def get_root(a, b, c):
    """ax² + bx + c = 0의 두 근을 반환"""
    d = b**2 - 4*a*c
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    return r1, r2   # 두 값을 튜플로 반환

# 개별 변수로 받기
root1, root2 = get_root(1, -3, 2)
print(f"root1 = {root1}, root2 = {root2}")

# 튜플 그대로 받기
roots = get_root(2, -6, -8)
print(f"튜플: {roots}  (type: {type(roots).__name__})")
````

### 심화 예제 — BMI 계산기

반환값을 활용하는 실용적인 예제입니다.


````python
def calc_bmi(weight_kg, height_cm):
    """BMI와 판정 결과를 반환"""
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    if bmi < 18.5:
        status = "저체중"
    elif bmi < 23:
        status = "정상"
    elif bmi < 25:
        status = "과체중"
    else:
        status = "비만"
    return round(bmi, 1), status

bmi_val, bmi_status = calc_bmi(70, 175)
print(f"BMI: {bmi_val}  →  {bmi_status}")

bmi_val, bmi_status = calc_bmi(50, 170)
print(f"BMI: {bmi_val}  →  {bmi_status}")
````

## 5.5 전역 변수와 지역 변수

| 구분 | 선언 위치 | 유효 범위 |
|------|-----------|-----------|
| **전역변수** | 함수 바깥 | 프로그램 전체 |
| **지역변수** | 함수 내부 | 해당 함수 내부만 |


````python
score = 100   # 전역변수

def show_score():
    score = 999  # 지역변수 (전역변수와 이름이 같지만 별개!)
    print(f"함수 내부 score: {score}")

show_score()
print(f"함수 외부 score: {score}")   # 전역변수는 그대로 100
````

````python
# 전역변수를 함수 내부에서 읽기 (참조)
total = 0

def add_to_total(value):
    # total을 읽기만 하면 오류 없음
    result = total + value
    return result

print(add_to_total(50))   # 50
print(f"total은 그대로: {total}")
````

````python
# global 키워드 — 전역변수를 함수 안에서 수정
counter = 0

def increment():
    global counter   # 전역변수임을 명시
    counter += 1

increment()
increment()
increment()
print(f"카운터: {counter}")   # 3
````

## 5.6 함수의 인자 전달 방식

### 5.6.1 디폴트 인자 (Default Argument)

매개변수에 기본값을 지정하면 인자 없이도 호출 가능합니다.  
> ⚠️ 디폴트 인자는 매개변수 목록의 **뒤쪽**에 위치해야 합니다.


````python
def greet(name, message="안녕하세요"):
    print(f"{message}, {name}님!")

greet("철수")               # 디폴트 메시지 사용
greet("영희", "반갑습니다") # 직접 메시지 전달
greet("민수", "좋은 아침")
````

````python
def power(base, exp=2):
    """base의 exp 제곱을 반환 (디폴트: 제곱)"""
    return base ** exp

print(power(3))     # 3² = 9  (디폴트 exp=2)
print(power(3, 3))  # 3³ = 27
print(power(2, 10)) # 2¹⁰ = 1024
````

### 5.6.2 키워드 인자 (Keyword Argument)

인자의 **이름을 명시**하여 순서에 관계없이 전달할 수 있습니다.


````python
def make_profile(name, age, city):
    print(f"이름: {name}, 나이: {age}, 도시: {city}")

# 위치 인자 (순서 중요)
make_profile("철수", 20, "서울")

# 키워드 인자 (순서 무관)
make_profile(age=25, city="부산", name="영희")

# 혼용: 위치 인자가 먼저!
make_profile("민수", city="대전", age=30)
````

````python
# 키워드 인자를 활용한 직사각형/원 넓이 계산기
def calc_area(shape, width=1, height=1, radius=1):
    if shape == "rect":
        return width * height
    elif shape == "circle":
        return 3.14159 * radius ** 2

rect_area   = calc_area("rect", width=5, height=3)
circle_area = calc_area("circle", radius=7)
print(f"직사각형 넓이: {rect_area}")
print(f"원 넓이: {circle_area:.2f}")
````

### 5.6.3 가변 인자 (Arbitrary Argument)

인자의 개수가 정해지지 않을 때 `*`를 사용합니다.  
가변 인자는 **튜플** 형태로 함수 내부에 전달됩니다.


````python
def sum_all(*numbers):
    """인자 개수에 상관없이 합계를 반환"""
    total = 0
    for n in numbers:
        total += n
    return total

print(sum_all(1, 2))              # 3
print(sum_all(1, 2, 3, 4, 5))    # 15
print(sum_all(10, 20, 30, 40))   # 100
````

````python
# 심화 — 가변 인자로 성적 통계 계산기
def grade_stats(subject, *scores):
    print(f"\n[{subject}]")
    print(f"  점수 목록: {scores}")
    print(f"  응시 인원: {len(scores)}명")
    print(f"  평균: {sum(scores)/len(scores):.1f}점")
    print(f"  최고: {max(scores)}점  최저: {min(scores)}점")

grade_stats("수학", 85, 92, 78, 95, 88)
grade_stats("영어", 70, 65, 80)
````

## 5.7 재귀함수 (Recursion)

재귀함수는 **자기 자신을 호출**하는 함수입니다.  
반드시 재귀 호출을 멈추는 **종료 조건(base case)**이 있어야 합니다.

### 팩토리얼 (n!)

```
n! = n × (n-1) × ... × 2 × 1
0! = 1 (종료 조건)
```


````python
def factorial(n):
    if n == 0:          # 종료 조건
        return 1
    return n * factorial(n - 1)   # 재귀 호출

for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

### 피보나치 수열

```
F(0) = 0, F(1) = 1
F(n) = F(n-1) + F(n-2)
```


````python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

print("피보나치 수열 (처음 15개):")
fib_list = [fibonacci(i) for i in range(15)]
print(fib_list)
````

````python
# 재귀 vs 반복문 — 두 방식 비교
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

n = 10
print(f"재귀:  fib({n}) = {fib_recursive(n)}")
print(f"반복:  fib({n}) = {fib_iterative(n)}")
````

## 🎯 종합 실습 — 성적 관리 프로그램

지금까지 배운 개념을 모두 활용한 미니 프로젝트입니다.

- 함수 정의 (`def`)
- 매개변수, 반환값 (`return`)
- 가변 인자 (`*args`)
- 디폴트 인자
- f-문자열 출력
- 내장 함수 (`min`, `max`, `sum`, `sorted`)


````python
def calc_stats(*scores):
    """점수 목록을 받아 통계를 딕셔너리로 반환"""
    n = len(scores)
    return {
        "count":   n,
        "total":   sum(scores),
        "average": sum(scores) / n,
        "highest": max(scores),
        "lowest":  min(scores),
        "sorted":  sorted(scores, reverse=True),
    }

def get_grade(score):
    """점수를 받아 학점(A~F)을 반환"""
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

def print_report(student_name, subject="전체", *scores):
    """성적 리포트를 출력하는 함수"""
    stats = calc_stats(*scores)
    grade = get_grade(stats["average"])

    print(f"\n{'='*40}")
    print(f"  성적 리포트  |  {student_name}  |  {subject}")
    print(f"{'='*40}")
    print(f"  응시 과목 수: {stats['count']}개")
    print(f"  총점:        {stats['total']}점")
    print(f"  평균:        {stats['average']:.1f}점  →  학점: {grade}")
    print(f"  최고점:      {stats['highest']}점")
    print(f"  최저점:      {stats['lowest']}점")
    print(f"  점수(고→저): {stats['sorted']}")
    print(f"{'='*40}")

# 실행
print_report("김철수", "이공계 과목", 88, 92, 75, 95, 80)
print_report("이영희", "인문계 과목", 70, 65, 88, 72)
````

## 📝 핵심 개념 정리

| 개념 | 키워드 / 문법 | 핵심 포인트 |
|------|--------------|------------|
| 함수 정의 | `def 이름():` | 코드 재사용, 블록화 |
| 매개변수 | `def f(a, b):` | 함수에 값 전달 |
| 반환값 | `return 값` | 함수 결과 돌려주기 |
| 다중 반환 | `return a, b` | 튜플로 반환 |
| 디폴트 인자 | `def f(n=1):` | 뒤쪽 매개변수에만 적용 |
| 키워드 인자 | `f(b=2, a=1)` | 위치 인자보다 뒤에 |
| 가변 인자 | `def f(*args):` | 튜플로 수집 |
| 재귀함수 | `f(n-1)` 호출 | 반드시 종료 조건 필요 |
| 전역변수 수정 | `global 변수` | 함수 내부에서 선언 |
| f-문자열 | `f"{변수:.2f}"` | format()보다 간결 |
