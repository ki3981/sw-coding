# Part 1. 🥄 Sequential (순차 구조)

> **컴퓨터는 코드를 위에서 아래로, 한 줄씩 실행합니다.**  
> 레시피를 ①→②→③ 순서대로 따라 하는 것과 같습니다.

```
① 냄비에 물 500ml 붓기
② 물 끓이기 🔥
③ 스프 넣기
④ 면 넣기
⑤ 3분 조리
⑥ 완성! 🍜
```


````python
# ── Sequential 예제: 라면 끓이기 ──────────────────────────────

print("=" * 42)
print("   🍜  라면 끓이기 — 순차 실행")
print("=" * 42)

# ① 재료 준비
water_ml  = 500
soup_pack = 1
noodle_g  = 100
print(f"① 재료 준비  →  물 {water_ml}ml / 스프 {soup_pack}봉 / 면 {noodle_g}g")

# ② 물 끓이기
temperature = 20
temperature = 100          # 값이 바뀜 (이전 값 20 → 100)
print(f"② 물 끓이기  →  현재 온도 {temperature}°C 🔥")

# ③ 스프 투입
soup_pack = soup_pack - 1
print(f"③ 스프 투입  →  남은 스프 {soup_pack}봉")

# ④ 면 투입
noodle_g = 0
print(f"④ 면  투입  →  남은 면 {noodle_g}g")

# ⑤ 조리 시간 계산
cook_min = 3
total_time = cook_min * 60   # 분 → 초 변환
print(f"⑤ 조리 중   →  {cook_min}분 ({total_time}초) ⏳")

# ⑥ 완성
calories = 500
print(f"⑥ 완성! 🍜  →  약 {calories} kcal")
print("=" * 42)
````

### ⚠️ 순서를 바꾸면 오류가 납니다

레시피에서 면을 넣기 전에 물을 끓여야 하듯,  
변수는 **반드시 선언한 뒤에** 사용해야 합니다.


````python
# ❌ 선언 전 사용 → NameError

try:
    total = ingredient_a + ingredient_b   # 아직 선언 안 됨!
except NameError as e:
    print(f"🚨 NameError: {e}")
    print("   → 재료(변수)를 먼저 준비(선언)해야 합니다!
")

# ✅ 올바른 순서
ingredient_a = 300
ingredient_b = 500
total = ingredient_a + ingredient_b
print(f"✅ 올바른 순서 → 합계: {total}원")
````

# Part 2. 🧂 Conditional (조건 구조)

> **"간이 싱거우면 소금을, 짜면 물을 더 넣는다"**  
> 조건에 따라 **다른 행동**을 수행합니다.

```
if 간이 싱겁다:
    소금을 더 넣는다
elif 간이 적당하다:
    그냥 둔다
else:
    물을 더 넣는다
```

### if / elif / else 구조
```python
if 조건A:
    # 조건A가 True일 때 실행
elif 조건B:
    # 조건B가 True일 때 실행
else:
    # 모두 아닐 때 실행
```


````python
# ── Conditional 예제 1: 국 간 맞추기 ──────────────────────────

saltiness = 3   # 간의 정도 (1=매우 싱거움 ~ 5=매우 짬)

print(f"현재 간 수치: {saltiness} / 5")
print("-" * 30)

if saltiness <= 2:
    print("🧂 간이 싱겁습니다 → 소금을 더 넣으세요!")
    saltiness += 1
    print(f"   소금 추가 후 간 수치: {saltiness}")
elif saltiness == 3:
    print("✅ 간이 딱 적당합니다 → 그대로 두세요!")
elif saltiness == 4:
    print("⚠️  간이 약간 짭니다 → 물을 조금 추가하세요.")
    saltiness -= 1
    print(f"   물 추가 후 간 수치: {saltiness}")
else:
    print("🚫 간이 너무 짭니다 → 물을 많이 추가하세요!")
    saltiness = 3
    print(f"   조절 후 간 수치: {saltiness}")
````

### 중첩 조건 (Nested if)

요리 중 **여러 조건을 동시에 확인**해야 할 때 사용합니다.  
예: 온도도 맞고 재료도 충분한지 함께 확인하기


````python
# ── Conditional 예제 2: 요리 가능 여부 판단 ───────────────────

temperature  = 100   # 냄비 온도 (°C)
stock_ml     = 300   # 남은 육수 (ml)
ingredient_ok = True  # 재료 준비 여부

print("🍲 찌개 조리 가능 여부 확인")
print(f"   온도: {temperature}°C  |  육수: {stock_ml}ml  |  재료: {'OK' if ingredient_ok else 'NG'}")
print("-" * 40)

if temperature >= 100:
    if stock_ml >= 200 and ingredient_ok:
        print("✅ 모든 조건 충족! → 찌개를 끓일 수 있습니다. 🥘")
    elif stock_ml < 200:
        print("⚠️  육수가 부족합니다 → 육수를 보충하세요.")
    else:
        print("⚠️  재료가 준비되지 않았습니다.")
else:
    print(f"🔥 온도가 낮습니다 ({temperature}°C) → 더 가열하세요.")
````

### 실전 예제: 요리 난이도 분류기


````python
# ── Conditional 예제 3: 요리 난이도 분류 ─────────────────────

recipes = [
    {"name": "라면",       "time": 5,  "skill": 1},
    {"name": "김치찌개",   "time": 30, "skill": 3},
    {"name": "스테이크",   "time": 45, "skill": 4},
    {"name": "마카롱",     "time": 90, "skill": 5},
]

print(f"{'요리':<10} {'시간':>6} {'스킬':>5}  난이도")
print("-" * 38)

for r in recipes:
    name, time, skill = r["name"], r["time"], r["skill"]

    if skill <= 1 and time <= 10:
        level = "⭐ 입문"
    elif skill <= 2 and time <= 20:
        level = "⭐⭐ 초급"
    elif skill <= 3 and time <= 40:
        level = "⭐⭐⭐ 중급"
    elif skill <= 4:
        level = "⭐⭐⭐⭐ 고급"
    else:
        level = "⭐⭐⭐⭐⭐ 전문가"

    print(f"{name:<10} {time:>4}분  {skill:>3}/5  {level}")
````

# Part 3. 🔁 Loop (반복 구조)

> **"3분 동안 계속 저어준다"** — 같은 동작을 반복합니다.

| 반복문 | 언제 사용? | 레시피 비유 |
|--------|-----------|------------|
| `for` | 횟수/목록이 정해진 반복 | "재료 5가지를 차례로 썬다" |
| `while` | 조건이 만족될 때까지 반복 | "물이 끓을 때까지 기다린다" |

### for 문 기본 구조
```python
for 변수 in 범위/리스트:
    # 반복할 코드
```

### while 문 기본 구조
```python
while 조건:
    # 조건이 True인 동안 반복
```


````python
# ── Loop 예제 1: for — 재료 손질하기 ─────────────────────────

ingredients = ["당근", "감자", "양파", "버섯", "호박"]

print("🔪 재료 손질 시작!")
print("-" * 30)

for i, item in enumerate(ingredients, start=1):
    print(f"  {i}번째: {item} 손질 완료 ✅")

print(f"
총 {len(ingredients)}가지 재료 손질 완료! 👍")
````

````python
# ── Loop 예제 2: for + range — 반죽 치대기 ──────────────────

knead_times = 10   # 치대는 횟수
elasticity  = 0    # 반죽 탄성 (0~100)

print(f"🍞 반죽 치대기 시작 (목표 {knead_times}회)")
print("-" * 35)

for i in range(1, knead_times + 1):
    elasticity += 8          # 한 번 칠 때마다 탄성 +8
    if i % 3 == 0:           # 3회마다 중간 점검
        print(f"  {i:>2}회 완료 → 탄성: {elasticity}  👀 점검")
    else:
        print(f"  {i:>2}회 완료 → 탄성: {elasticity}")

print(f"
최종 반죽 탄성: {elasticity} / 100 🎉")
````

### `while` — 조건이 만족될 때까지 반복


````python
# ── Loop 예제 3: while — 물이 끓을 때까지 ───────────────────

temperature = 20   # 초기 온도 (°C)
target_temp = 100  # 목표 온도

print(f"🔥 물 끓이기 시작! (현재: {temperature}°C → 목표: {target_temp}°C)")
print("-" * 42)

step = 0
while temperature < target_temp:
    temperature += 20    # 1단계마다 20°C 상승
    step += 1
    status = "펄펄 끓음! 🫧" if temperature >= target_temp else "가열 중..."
    print(f"  [{step}단계] {temperature:>3}°C  {status}")

print(f"
✅ {step}단계 만에 목표 온도 도달! 면을 넣어도 됩니다.")
````

### `break` & `continue` — 조리 중 예외 처리

| 키워드 | 비유 |
|--------|------|
| `break` | "탄 냄새가 나면 즉시 불 끔" |
| `continue` | "상한 재료는 건너뛰고 다음 재료로" |


````python
# ── Loop 예제 4: break & continue ────────────────────────────

vegetables = ["당근", "상한배추", "양파", "썩은버섯", "호박", "불량두부"]
good_count = 0

print("🥬 재료 품질 검사 시작!")
print("-" * 35)

for veg in vegetables:
    # 상하거나 썩은 재료는 건너뜀 (continue)
    if "상한" in veg or "썩은" in veg:
        print(f"  ⚠️  {veg} → 불량, 건너뜁니다 (continue)")
        continue

    # 불량 재료 발견 시 전체 조리 중단 (break)
    if "불량" in veg:
        print(f"  🚫 {veg} → 심각한 불량! 조리 중단 (break)")
        break

    good_count += 1
    print(f"  ✅ {veg} → 사용 가능 ({good_count}번째)")

print(f"
최종 사용 가능 재료: {good_count}가지")
````

# Part 4. 🍱 통합 예제: 도시락 주문 시스템

> Sequential + Conditional + Loop 을 모두 활용합니다.

**시나리오**: 여러 명의 주문을 받아 총 금액을 계산하고,  
주문 금액에 따라 할인을 적용합니다.


````python
# ── 통합 예제: 도시락 주문 & 할인 시스템 ────────────────────

# ① Sequential: 메뉴 데이터 준비
menu = {
    "불고기 도시락": 8500,
    "제육 도시락":   7500,
    "김치 도시락":   6500,
    "채소 도시락":   6000,
}

orders = [
    {"name": "김철수", "item": "불고기 도시락", "qty": 2},
    {"name": "이영희", "item": "제육 도시락",   "qty": 1},
    {"name": "박민준", "item": "김치 도시락",   "qty": 3},
    {"name": "최수진", "item": "채소 도시락",   "qty": 5},
]

print("=" * 50)
print("   🍱  도시락 주문 처리 시스템")
print("=" * 50)

total_revenue = 0   # 총 매출

# ② Loop: 모든 주문을 순서대로 처리
for order in orders:
    customer = order["name"]
    item     = order["item"]
    qty      = order["qty"]
    price    = menu[item]         # Sequential: 단가 조회
    subtotal = price * qty        # Sequential: 소계 계산

    # ③ Conditional: 수량에 따른 할인 적용
    if qty >= 5:
        discount_rate = 0.10      # 5개 이상 → 10% 할인
    elif qty >= 3:
        discount_rate = 0.05      # 3개 이상 → 5% 할인
    else:
        discount_rate = 0.0       # 할인 없음

    discount = int(subtotal * discount_rate)
    final    = subtotal - discount

    # Sequential: 결과 출력
    discount_str = f"(-{discount:,}원 {int(discount_rate*100)}%↓)" if discount > 0 else ""
    print(f"  {customer:<6}  {item}  x{qty}  →  {final:>7,}원  {discount_str}")

    total_revenue += final   # Sequential: 누적 매출

print("-" * 50)
print(f"  💰 총 매출: {total_revenue:,}원")

# ③ Conditional: 매출 등급 판정
print()
if total_revenue >= 100000:
    grade = "🥇 골드"
    bonus = "사장님 보너스 지급!"
elif total_revenue >= 60000:
    grade = "🥈 실버"
    bonus = "내일도 화이팅!"
else:
    grade = "🥉 브론즈"
    bonus = "홍보가 필요합니다."

print(f"  오늘의 매출 등급: {grade}")
print(f"  한마디: {bonus}")
print("=" * 50)
````

# Part 5. ✅ 핵심 정리

## 세 가지 제어 구조 비교

| 구조 | 흐름 | 레시피 비유 | 키워드 |
|------|------|------------|--------|
| **Sequential** | 위→아래 순서 실행 | 레시피 단계 순서대로 | (없음, 기본 흐름) |
| **Conditional** | 조건에 따라 분기 | "싱거우면 소금 추가" | `if / elif / else` |
| **Loop** | 조건/횟수만큼 반복 | "3분간 계속 저어라" | `for / while` |

## 반복문 비교

| | `for` | `while` |
|-|-------|---------|
| **언제** | 횟수·목록이 정해진 경우 | 조건이 만족될 때까지 |
| **비유** | 재료 5개를 차례로 썬다 | 물이 끓을 때까지 기다린다 |
| **주의** | 범위 실수 (off-by-one) | 무한루프 주의 |

> 🍳 **"좋은 코드는 좋은 레시피처럼, 순서가 명확하고 예외 처리가 잘 되어 있어야 합니다."**
