# 🐍 Python 제어문: 조건문과 반복문
### 조건문 (3장) + 반복문 (4장) — 단계적 학습 노트북

이 노트북은 **하나의 간단한 코드**에서 시작해서, 조건문과 반복문을 차례로 추가하며 Python 제어문의 흐름을 배웁니다.

---
## 📌 STEP 0: 출발점 — 순차문 (Sequential Statements)

프로그램은 기본적으로 **위에서 아래로** 순서대로 실행됩니다.  
아래 코드는 학생의 점수를 변수에 저장하고 출력하는 가장 단순한 형태입니다.

````python
# 학생 점수 — 순차적으로 실행되는 코드
score = 75
print("점수:", score)
````

---
## 📌 STEP 1: `if` 조건문 추가 — 합격 여부 판별

> **조건문**은 특정 조건이 참(True)일 때만 코드를 실행합니다.  
> 점수가 60점 이상이면 '합격'을 출력해 봅시다.

```
if 조건식:
    실행할 코드  # ← 반드시 들여쓰기 (스페이스 4칸)
```

````python
score = 75
print("점수:", score)

# ✅ 추가된 부분: if 조건문
if score >= 60:
    print("합격")
````

💡 **확인해 보세요:** `score = 45`로 바꾸면 어떻게 될까요?

---
## 📌 STEP 2: `if-else` 추가 — 합격 / 불합격 구분

> 조건이 거짓(False)일 때 실행할 코드도 지정할 수 있습니다.  
> **배타적 관계**: 둘 중 하나만 반드시 실행됩니다.

```
if 조건식:
    조건이 참일 때 실행
else:
    조건이 거짓일 때 실행
```

````python
score = 75
print("점수:", score)

if score >= 60:
    print("합격")
# ✅ 추가된 부분: else
else:
    print("불합격")
````

---
## 📌 STEP 3: `if-elif-else` — 학점 등급 산출기

> 조건이 여러 개일 때 `elif`를 사용합니다.  
> 위에서부터 순서대로 검사하고, **처음 True인 블록만** 실행합니다.

| 점수 | 학점 |
|------|------|
| 90 이상 | A |
| 80 이상 | B |
| 70 이상 | C |
| 60 이상 | D |
| 60 미만 | F |

````python
score = 75
print("점수:", score)

# ✅ 추가된 부분: elif로 여러 조건 처리
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("학점:", grade)
````

💡 **확인해 보세요:** `score = 88`, `score = 55`로 바꿔서 결과를 확인해 보세요.

---
## 📌 STEP 4: `and` / `or` 복합 조건식

> 두 가지 이상의 조건을 동시에 검사할 때 논리 연산자를 사용합니다.

| 연산자 | 의미 | 결과 |
|--------|------|------|
| `and` | 둘 다 참이어야 True | A and B |
| `or`  | 하나라도 참이면 True | A or B |

점수가 70점 이상이고, 출석률이 80% 이상일 때 '통과'를 판별해 봅니다.

````python
score = 75
attendance = 85   # 출석률 (%)

print("점수:", score)
print("출석률:", attendance)

# ✅ 추가된 부분: and 복합 조건
if score >= 70 and attendance >= 80:
    print("✅ 통과: 점수와 출석률 모두 기준 충족")
else:
    print("❌ 미통과: 점수 또는 출석률이 기준 미달")
````

💡 **확인해 보세요:** `attendance = 60`으로 바꾸면 어떻게 될까요?

---
## 📌 STEP 5: `for` 반복문 — 여러 학생 점수 처리

> **반복문**은 같은 코드를 여러 번 실행할 때 사용합니다.  
> `for`문은 반복 횟수가 정해진 경우에 적합합니다.

```
for 변수 in range(횟수):
    반복할 코드
```

학생이 5명이라면, 같은 코드를 5번 반복해야 합니다.

````python
# ✅ 추가된 부분: for 반복문으로 5번 반복 출력
for i in range(5):
    score = 75        # 일단 모든 학생의 점수를 같다고 가정
    print(f"학생 {i+1} | 점수: {score}")
````

---
## 📌 STEP 6: `for` + 리스트 — 실제 점수 목록 처리

> `range()` 대신 **리스트**를 직접 사용할 수 있습니다.  
> 리스트의 원소를 순서대로 하나씩 꺼내 처리합니다.

````python
# 학생 5명의 실제 점수 리스트
scores = [75, 92, 58, 83, 67]

# ✅ 추가된 부분: 리스트를 for문으로 순회
for i, score in enumerate(scores):
    print(f"학생 {i+1} | 점수: {score}")
````

---
## 📌 STEP 7: `for` + `if` 결합 — 합격 여부 자동 판별

> 반복문 안에 조건문을 넣어 **각 학생의 합격 여부를 자동으로 판별**합니다.  
> 앞서 배운 두 가지 제어문이 함께 사용됩니다.

````python
scores = [75, 92, 58, 83, 67]

for i, score in enumerate(scores):
    # ✅ 추가된 부분: 반복문 안에 조건문
    if score >= 60:
        result = "합격 ✅"
    else:
        result = "불합격 ❌"
    print(f"학생 {i+1} | 점수: {score} | {result}")
````

---
## 📌 STEP 8: `for` + 누적 합산 — 전체 평균 계산

> 반복문을 이용해 변수에 값을 **누적**하면 합계와 평균을 구할 수 있습니다.  
> `total = total + score` → 루프를 돌 때마다 누적됩니다.

````python
scores = [75, 92, 58, 83, 67]

# ✅ 추가된 부분: 누적 합산으로 평균 계산
total = 0
for score in scores:
    total = total + score

average = total / len(scores)
print(f"전체 합계: {total}점")
print(f"전체 평균: {average:.1f}점")
````

---
## 📌 STEP 9: `while` 반복문 — 조건 기반 반복

> `while`문은 반복 횟수를 모르지만 **조건이 명확할 때** 사용합니다.  
> 평균 점수 이상인 학생 수를 셀 때, 몇 명인지 미리 모를 수 있습니다.

```
while 조건식:
    조건이 참인 동안 반복
```

````python
scores = [75, 92, 58, 83, 67]
average = sum(scores) / len(scores)

# ✅ 추가된 부분: while 반복문
i = 0
above_avg_count = 0

while i < len(scores):
    if scores[i] >= average:
        above_avg_count += 1
    i += 1   # ⚠️ 반드시 증가! 없으면 무한 루프

print(f"평균: {average:.1f}점")
print(f"평균 이상인 학생 수: {above_avg_count}명")
````

---
## 📌 STEP 10: `break` & `continue` — 반복 흐름 제어

> 반복문 안에서 흐름을 바꿔야 할 때 사용합니다.

| 키워드 | 동작 |
|--------|------|
| `break` | 반복문을 **즉시 종료** |
| `continue` | 이번 루프의 나머지를 **건너뛰고** 다음 반복으로 이동 |

**break 예시:** 점수가 60점 미만인 첫 번째 학생을 발견하면 탐색을 멈춥니다.

````python
scores = [75, 92, 58, 83, 67]

# ✅ break: 불합격 학생 발견 즉시 중단
print("--- break 예시 ---")
for i, score in enumerate(scores):
    if score < 60:
        print(f"학생 {i+1}이 불합격({score}점)! 탐색 중단.")
        break
    print(f"학생 {i+1}: {score}점 — 합격")
````

````python
scores = [75, 92, 58, 83, 67]

# ✅ continue: 불합격 학생은 건너뛰고 합격자만 출력
print("--- continue 예시 ---")
for i, score in enumerate(scores):
    if score < 60:
        continue   # 60점 미만이면 아래 print를 건너뜀
    print(f"학생 {i+1}: {score}점 — 합격 ✅")
````

## 🏁 최종 완성 코드 — 모든 개념 통합

지금까지 배운 **조건문 + 반복문**을 모두 활용해  
학생 성적 처리 프로그램을 완성합니다.

````python
# ============================================
# 📋 학생 성적 처리 프로그램 (완성본)
# ============================================

scores = [75, 92, 58, 83, 67]
names  = ["김민준", "이서연", "박지호", "최수아", "정도윤"]

print("=" * 40)
print("        📊 학생 성적 처리 결과")
print("=" * 40)

total = 0
pass_count = 0

for i, score in enumerate(scores):
    total += score

    # 학점 산출 (if-elif-else)
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"

    # 합격 여부 (if-else)
    if grade != "F":
        result = "합격 ✅"
        pass_count += 1
    else:
        result = "불합격 ❌"

    print(f"  {names[i]:5s} | 점수: {score:3d} | 학점: {grade} | {result}")

print("-" * 40)
average = total / len(scores)
print(f"  평균 점수: {average:.1f}점")
print(f"  합격 인원: {pass_count}/{len(scores)}명")
print("=" * 40)
````

---
## 📝 핵심 정리

| 제어문 | 키워드 | 사용 시점 |
|--------|--------|----------|
| 조건문 | `if` / `elif` / `else` | 조건에 따라 다른 코드를 실행할 때 |
| 정해진 반복 | `for` + `range()` / 리스트 | 반복 횟수가 명확할 때 |
| 조건 기반 반복 | `while` | 반복 조건만 명확할 때 |
| 반복 종료 | `break` | 특정 조건에서 반복문 탈출 |
| 반복 건너뜀 | `continue` | 특정 조건의 반복 회차만 건너뜀 |

> ⚠️ **들여쓰기 규칙**: Python에서 들여쓰기(스페이스 4칸)는 코드 블록을 구분하는 문법입니다. 반드시 일관되게 사용하세요!
